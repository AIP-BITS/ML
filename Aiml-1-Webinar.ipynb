### Regression

```python
import matplotlib.pyplot as plt
import numpy as np

# Define two data points
x1, y1 = 1, 2
x2, y2 = 3, 4

# Calculate the slope (m)
m = (y2 - y1) / (x2 - x1)

# Calculate the intercept (c)
c = y1 - m * x1

# Regression line equations
print(f"Regression Line Equation: y = {m}x + {c}")

# Plot the data points
plt.scatter([x1, x2], [y1, y2], color='red', label='Data Points')

# Plot the regression line
x_line = np.linspace(min(x1, x2) - 1, max(x1, x2) + 1, 100)  # Extend line slightly beyond points
y_line = m * x_line + c
plt.plot(x_line, y_line, color='blue', label=f'y = {m}x + {c}')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regression Line for Two Points')
plt.legend()

# Show plot
plt.show()
```

### Classification

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# Dataset with two features and target class
data = [
    (3, 5, "Yes"),  # (servings, yawns, class)
    (2, 3, "No"),
    (4, 4, "Yes"),
    (1, 2, "No"),
    (5, 6, "Yes"),
]

# New data point to classify
new_point = (3, 4)

# Calculate Euclidean distance
def euclidean_distance(point1, point2):
    return np.sqrt((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2)

# Compute distances to all points
distances = [(euclidean_distance(new_point, (x, y)), label) for x, y, label in data]

# Get the 3 nearest neighbors
nearest_neighbors = sorted(distances)[:3]
```

```python
mode = {}
for i,j in nearest_neighbors:
    if j in mode:
        mode[j] = mode[j]+1
    else:
        mode[j]=1
```

```python
max_key = max(mode, key=mode.get)
max_key
```

```python

# Predict the class based on the majority vote
predicted_class = Counter([label for _, label, _ in nearest_neighbors]).most_common(1)[0][0]

# Visualization
plt.figure(figsize=(8, 6))

# Plot the dataset
for x, y, label in data:
    color = 'green' if label == "Yes" else 'red'
    plt.scatter(x, y, color=color)

# Highlight the new point
plt.scatter(new_point[0], new_point[1], color='blue', s=100, label="New Point")


# Add labels and legend
plt.xlabel("Number of Servings")
plt.ylabel("Number of Yawns")
plt.title(f"Prediction: {predicted_class}")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
