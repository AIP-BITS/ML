```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
movie_data = pd.read_csv('data/movie_dataset.csv', header = None)
num_records = len(movie_data)
print(num_records)
```

```python
movie_data
```

```python
records = []
for i in range(0, num_records):
    records.append([str(movie_data.values[i,j]) for j in range(0, 20)])
```

```python
# -Support
# -confidence
# -Lift 
```

```python
association_rules = apriori(records, min_support=0.0053, min_confidence=0.20, min_lift=3, min_length=2)
association_results = list(association_rules)
```
